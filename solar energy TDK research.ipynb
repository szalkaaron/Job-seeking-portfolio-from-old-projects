Analysis of Hungarian electricity chaneges on net export 2018-2023

This project was part of a TDK research (University scientific cometition) that I wrote with a co-author, László Németh. He wrote the literature reveiew in the beginning and I did the data analysis and empirical research part. I wrote this code below to understand the dataset and to make different plots for the research paper. I downloaded the data ('solar energy TDK research.csv') from Mavir's webpage, that cotains electricity usage data for every 15 minets for 5 years making 175295 lines. Our goal was to understand more the effect of increased volumes of solar energy on the volatility of the net export of eletricity.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.patches as mpatches
```

```python
df = pd.read_csv(r'solar energy TDK research.csv', sep = ';', header=0)

my_list = []
my_list2 = []
my_list3 = []
my_list4 = []

hanynap = int(175296/96 + 1)
hanyhet = int(175296/(96*7) + 1)
hanyhónap = int(175296/(96*30) + 1)
hanyév = int(175296/(96*365) + 1)


for i in range(1, hanynap):
    my_list.append(i)

for i in range(1, hanyhet): 
    my_list2.append(i)  

for i in range(1, hanyhónap): 
    my_list3.append(i)  

for i in range(1, hanyév): 
    my_list4.append(i)  
```

```python
data = pd.DataFrame()
data['Nap'] = my_list

dataheti = pd.DataFrame()
dataheti['Hét'] = my_list2

datahavi = pd.DataFrame()
datahavi['Hónap'] = my_list3

dataéves = pd.DataFrame()
dataéves['Év'] = my_list4
```

```python
#HETI

num_hetek = dataheti.shape[0]
darab = 96*7


imphetisumok = []
for i in range(0, num_rows, 672):
    imphetisum = df['Import-export szaldó - üzemirányítási (15 perces)'].iloc[i:i+672].sum()
    imphetisumok.append(imphetisum)
imphetisumok.pop(-1)

imphetistdk = []
for i in range(0, num_rows, 672):
    imphetistd = df['Import-export szaldó - üzemirányítási (15 perces)'].iloc[i:i+672].std()
    imphetistdk.append(imphetistd)
imphetistdk.pop(-1)

naparanyhetik = []
for i in range(0, num_rows, 672):
    naparanyheti = (df['Naperőművek net.üzemir.mérés (15p)'].iloc[i:i+672].sum()/df['Nettó rendszerterhelés tény - üzemirányítási (15 perc)'].iloc[i:i+672].sum())
    naparanyhetik.append(naparanyheti)
naparanyhetik.pop(-1)

hetiosszestermelesek = []
for i in range(0, num_rows, darab):
    hetiosszestermeles = df['Hazai termelés (erőművi szumma) net.üzemir.mérés (15p)'].iloc[i:i+darab].mean()
    hetiosszestermelesek.append(hetiosszestermeles)
hetiosszestermelesek.pop(-1)
```

```python
#HAVI

num_honapok = datahavi.shape[0]
darab = 96*30


imphavisumok = []
for i in range(0, num_rows, darab):
    imphavisum = df['Import-export szaldó - üzemirányítási (15 perces)'].iloc[i:i+darab].sum()
    imphavisumok.append(imphavisum)
imphavisumok.pop(-1)

imphavistdk = []
for i in range(0, num_rows, darab):
    imphavistd = df['Import-export szaldó - üzemirányítási (15 perces)'].iloc[i:i+darab].std()
    imphavistdk.append(imphavistd)
imphavistdk.pop(-1)

naparanyhavik = []
for i in range(0, num_rows, darab):
    naparanyhavi = (df['Naperőművek net.üzemir.mérés (15p)'].iloc[i:i+darab].sum()/df['Nettó rendszerterhelés tény - üzemirányítási (15 perc)'].iloc[i:i+darab].sum())
    naparanyhavik.append(naparanyhavi)
naparanyhavik.pop(-1)

haviosszestermelesek = []
for i in range(0, num_rows, darab):
    haviosszestermeles = df['Hazai termelés (erőművi szumma) net.üzemir.mérés (15p)'].iloc[i:i+darab].mean()
    haviosszestermelesek.append(haviosszestermeles)
haviosszestermelesek.pop(-1)
```

```python
#ÉVES

num_évek = dataéves.shape[0]
darab = 96*365


impévisumok = []
for i in range(0, num_rows, darab):
    impévisum = df['Import-export szaldó - üzemirányítási (15 perces)'].iloc[i:i+darab].sum()
    impévisumok.append(impévisum)
impévisumok.pop(-1)

impévistdk = []
for i in range(0, num_rows, darab):
    impévistd = df['Import-export szaldó - üzemirányítási (15 perces)'].iloc[i:i+darab].std()
    impévistdk.append(impévistd)
impévistdk.pop(-1)

naparanyévik = []
for i in range(0, num_rows, darab):
    naparanyévi = (df['Naperőművek net.üzemir.mérés (15p)'].iloc[i:i+darab].sum()/df['Nettó rendszerterhelés tény - üzemirányítási (15 perc)'].iloc[i:i+darab].sum())
    naparanyévik.append(naparanyévi)
naparanyévik.pop(-1)

éviosszestermelesek = []
for i in range(0, num_rows, darab):
    éviosszestermeles = df['Hazai termelés (erőművi szumma) net.üzemir.mérés (15p)'].iloc[i:i+darab].sum()
    éviosszestermelesek.append(éviosszestermeles)
éviosszestermelesek.pop(-1)

éviosszesnapenergianélkülitermelesek = []
for i in range(0, num_rows, darab):
    éviosszesnapenergianélkülitermeles = df['Hazai termelés (erőművi szumma) net.üzemir.mérés (15p)'].iloc[i:i+darab].sum()
    éviosszesnapenergianélkülitermelesek.append(éviosszesnapenergianélkülitermeles)
éviosszesnapenergianélkülitermelesek.pop(-1)

éviosszesnapenergiatermelesek = []
for i in range(0, num_rows, darab):
    éviosszesnapenergiatermeles = df['Naperőművek net.üzemir.mérés (15p)'].iloc[i:i+darab].sum()
    éviosszesnapenergiatermelesek.append(éviosszesnapenergiatermeles)
éviosszesnapenergiatermelesek.pop(-1)

éviosszesnapenergiafogyasztások = []
for i in range(0, num_rows, darab):
    éviosszesnapenergiafogyasztás = df['Nettó rendszerterhelés tény - üzemirányítási (15 perc)'].iloc[i:i+darab].sum()
    éviosszesnapenergiafogyasztások.append(éviosszesnapenergiafogyasztás)
éviosszesnapenergiafogyasztások.pop(-1)
```

```python
data['Napi napenergia átlag'] = averages
data['Napi import std'] = impstds
data['Napi nettó import variancia'] = impvars
data['Napi napenergia aránya'] = naparanyok
data['Napi energiatermelés átlag'] = napiosszestermelesek

dataheti['Heti import sum'] = imphetisumok
dataheti['Heti import std'] = imphetistdk
dataheti['Heti napenergia aránya'] = naparanyhetik
dataheti['Heti energiatermelés átlag'] = hetiosszestermelesek

datahavi['Havi import sum'] = imphavisumok
datahavi['Havi import std'] = imphavistdk
datahavi['Havi napenergia aránya'] = naparanyhavik
datahavi['Havi energiatermelés átlag'] = haviosszestermelesek

dataéves['Évi import sum'] = impévisumok
dataéves['Évi import std'] = impévistdk
dataéves['Évi napenergia aránya'] = naparanyévik
dataéves['Évi energiatermelés összeg'] = éviosszestermelesek
dataéves['Évi napenergiatermelés összeg'] = éviosszesnapenergiatermelesek
dataéves['Évi energiatermelés összeg napenergia nélkül'] = éviosszesnapenergianélkülitermelesek
dataéves['Évi energiafogyasztás összeg'] = éviosszesnapenergiafogyasztások
```
